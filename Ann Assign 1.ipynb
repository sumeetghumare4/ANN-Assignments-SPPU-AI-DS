***1. Write a Python program to plot a few activation functions that are being used in neural networks.***

```python
import numpy as np
import matplotlib.pyplot as plt
from derivative import dxdt
```

**1. Sigmoid Activation Function**

```python
x = np.linspace(-20,20,100)
sig = np.linspace(-20,20,100)
sig= 1/(1 + np.exp(-x))
plt.plot(x,sig)
```

**Derivative Sigmoid**

```python
dydx = dxdt(sig,x)
plt.plot(x,dydx)
```

**2. TanH Activation Function**

```python
x = np.linspace(-20,20,100)
ta = np.linspace(-20,20,100)
ta = (np.exp(ta)-np.exp(-ta))/(np.exp(ta)+np.exp(-ta))
plt.plot(x,ta)
```

**Derivative Tanh**

```python
tadydx = dxdt(ta,x)
plt.plot(x ,tadydx)
```

**3. Relu Activation Function**

```python
x = np.linspace(-20,20,100)
rel = x
rel = np.maximum(0,rel)
plt.plot(x, rel)
```

**Derivative Relu**

```python
reldydx = dxdt(rel ,x)
plt.plot(x, reldydx)
```

**4. Leaky Relu Activation Function**

```python
x = np.linspace(-20,20,100)
le_rel = np.maximum(x*0.1,x)
plt.plot(x,le_rel)
```

**Derivative Leaky Relu**

```python
le_reldydx = dxdt(le_rel , x)
plt.plot(x, le_reldydx)
```

**5. Softplus Activation Function**

```python
x = np.linspace(-20,20,100)
soft = np.log(1+(np.exp(x)))
plt.plot(x, soft)
```

**Derivative SoftPlus**

```python
softdydx = dxdt(soft ,x)
plt.plot(x,softdydx)
```
